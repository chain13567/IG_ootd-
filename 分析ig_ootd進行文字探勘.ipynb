```python
import jieba
import csv
import operator
from collections import Counter
import pandas as pd
```

```python
jieba.set_dictionary('/content/drive/MyDrive/Colab Notebooks/project/åˆ†æOOTDæ–‡å­—æ¢å‹˜/dict.txt.big.txt')
jieba.load_userdict('/content/drive/MyDrive/Colab Notebooks/project/åˆ†æOOTDæ–‡å­—æ¢å‹˜/user_dic.txt')
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/project/åˆ†æOOTDæ–‡å­—æ¢å‹˜/stop_new.txt') as f:
    stops = f.read().split('\n')
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/project/åˆ†æOOTDæ–‡å­—æ¢å‹˜/å…§å®¹.csv', encoding='utf-8-sig') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        row = str(row)
        #åˆªé™¤ç„¡æ„ç¾©å­—
        row = row.replace(' ','')
        row = row.replace('.','')
        row = row.replace('_','')
        row = row.replace('__','')
        row = row.replace('-','')
        row = row.replace('&','')
        row = row.replace('â€™','')
        corpus = []                # corpus èªæ–™åº«
        # é–‹å§‹çµjiebaåˆ†è©ä¸¦å»é™¤ä¸è¦çš„å­—
        # cut_all=False ç²¾ç¡®æ¨¡å¼
        corpus += [t for t in jieba.cut(row, cut_all=False) if t not in stops]
        # å‰µå€‹ç©ºçš„dictionary
        dic = {}
        for ele in corpus:
            if ele not in dic:
                dic[ele] = 1
            else:
                dic[ele] = dic[ele] + 1
        # é¡¯ç¤ºåˆ‡å­—çš„çµæœ
        sorted_word = sorted(dic.items(), key=operator.itemgetter(1), reverse=True)
        print(sorted_word)
        with open('te.csv', 'a', encoding='utf-8-sig') as wf2:
            for ele in sorted_word:
                # é€™è¡Œæ˜¯è¡¨ç¤ºæˆ‘è¦2å€‹å­—ä»¥ä¸Šçš„å¥å­
                if len(ele[0]) >= 2:
                    world = ele[0] + ','
                    wf2.write(world)
```

```python
house = open('te.csv', encoding='utf-8-sig').read()
# å°‡è®€å–çš„æª”æ¡ˆå…§ä¹‹å­—å–®ç¨åˆ†é–‹
list1 = house.split(',')
# è¨ˆç®—å‡ºç¾æ¬¡æ•¸
dir1 = Counter(list1)
```

```python
# å°‡å€¼å¯«å…¥csvæª”ä¸­
with open('te1.csv', encoding='utf-8-sig', mode='w') as fp:
    fp.write('é—œéµå­—,å€¼,\n')
    for tag, count in dir1.items():
        fp.write('{},{}\n'.format(tag, count))
```
