```python
import jieba
import csv
import operator
from collections import Counter
import pandas as pd
```

```python
jieba.set_dictionary('/content/drive/MyDrive/Colab Notebooks/project/分析OOTD文字探勘/dict.txt.big.txt')
jieba.load_userdict('/content/drive/MyDrive/Colab Notebooks/project/分析OOTD文字探勘/user_dic.txt')
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/project/分析OOTD文字探勘/stop_new.txt') as f:
    stops = f.read().split('\n')
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/project/分析OOTD文字探勘/內容.csv', encoding='utf-8-sig') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        row = str(row)
        corpus = []                # corpus 語料庫
        # 開始結jieba分詞並去除不要的字
        # cut_all=False 精确模式
        corpus += [t for t in jieba.cut(row, cut_all=False) if t not in stops]
        # 創個空的dictionary
        dic = {}
        for ele in corpus:
            if ele not in dic:
                dic[ele] = 1
            else:
                dic[ele] = dic[ele] + 1
        # 顯示切字的結果
        sorted_word = sorted(dic.items(), key=operator.itemgetter(1), reverse=True)
        print(sorted_word)
        with open('te.csv', 'a', encoding='utf-8-sig') as wf2:
            for ele in sorted_word:
                # 這行是表示我要2個字以上的句子
                if len(ele[0]) >= 2:
                    world = ele[0] + ','
                    wf2.write(world)
```

```python
house = open('te.csv', encoding='utf-8-sig').read()
# 將讀取的檔案內之字單獨分開
list1 = house.split(',')
# 計算出現次數
dir1 = Counter(list1)
```

```python
print(list1)
```

```python
print(dir1)
```

```python
# 將值寫入csv檔中
with open('te1.csv', encoding='utf-8-sig', mode='w') as fp:
    fp.write('關鍵字,值,\n')
    for tag, count in dir1.items():
        fp.write('{},{}\n'.format(tag, count))
```
