```python
import jieba
import csv
import operator
from collections import Counter
import pandas as pd
```

```python
jieba.set_dictionary('/content/drive/MyDrive/Colab Notebooks/project/分析OOTD文字探勘/dict.txt.big.txt')
jieba.load_userdict('/content/drive/MyDrive/Colab Notebooks/project/分析OOTD文字探勘/user_dic.txt')
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/project/分析OOTD文字探勘/stop_new.txt') as f:
    stops = f.read().split('\n')
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/project/分析OOTD文字探勘/內容.csv', encoding='utf-8-sig') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        row = str(row)
        # 刪除無意義字
        row = row.replace(' ','')
        row = row.replace('.','')
        row = row.replace('_','')
        row = row.replace('__','')
        row = row.replace('-','')
        row = row.replace('&','')
        row = row.replace('’','')
        # 將英文統一轉小寫
        row = row.replace('A','a')
        row = row.replace('B','b')
        row = row.replace('C','c')
        row = row.replace('D','d')
        row = row.replace('E','e')
        row = row.replace('F','f')
        row = row.replace('G','g')
        row = row.replace('H','h')
        row = row.replace('I','i')
        row = row.replace('J','j')
        row = row.replace('K','k')
        row = row.replace('L','l')
        row = row.replace('M','m')
        row = row.replace('N','n')
        row = row.replace('O','o')
        row = row.replace('P','p')
        row = row.replace('Q','q')
        row = row.replace('R','r')
        row = row.replace('S','s')
        row = row.replace('T','t')
        row = row.replace('U','u')
        row = row.replace('V','v')
        row = row.replace('W','w')
        row = row.replace('X','x')
        row = row.replace('Y','y')
        row = row.replace('Z','z')
        corpus = []                # corpus 語料庫
        # 開始結jieba分詞並去除不要的字
        # cut_all=False 精确模式
        corpus += [t for t in jieba.cut(row, cut_all=False) if t not in stops]
        # 創個空的dictionary
        dic = {}
        for ele in corpus:
            if ele not in dic:
                dic[ele] = 1
            else:
                dic[ele] = dic[ele] + 1
        # 顯示切字的結果
        sorted_word = sorted(dic.items(), key=operator.itemgetter(1), reverse=True)
        print(sorted_word)
        with open('te.csv', 'a', encoding='utf-8-sig') as wf2:
            for ele in sorted_word:
                # 這行是表示我要2個字以上的句子
                if len(ele[0]) >= 2:
                    world = ele[0] + ','
                    wf2.write(world)
```

```python
house = open('te.csv', encoding='utf-8-sig').read()
# 將讀取的檔案內之字單獨分開
list1 = house.split(',')
# 計算出現次數
dir1 = Counter(list1)
```

```python
# 將值寫入csv檔中
with open('te1.csv', encoding='utf-8-sig', mode='w') as fp:
    fp.write('關鍵字,值,\n')
    for tag, count in dir1.items():
        fp.write('{},{}\n'.format(tag, count))
```
